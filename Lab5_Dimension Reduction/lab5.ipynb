Пока можно удалить признаки-дефекты (при чтении csv файла как-то считались пустые колонки без данных) + колонки с очень маленьким количеством объектов относительно всей выборки

```python
df = df.drop(columns=['Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', "W'W'", 'ff10'])
```

В некоторых строчках колонки `c` остались какие-то дефекты html разметки => удалим эти строки, также переименуем колонки и закодируем время в формате тригонометрических значений для сохранения информации о дистанции и зацикленностей дней и месяцев относительно друг другапе

```python
# drop rows that are bugged
df['c'] = df['c'].apply(lambda s: s.split()[0])
df = df.drop(df[df['c']=='<span'].index)
```

```python
# encode time
df = df.rename(columns={'Местное время в Вене / Швехате (аэропорт)':"Time"})

df['day'] = df['Time'].apply(lambda x: int(x.split('.')[0]))
df['month'] = df['Time'].apply(lambda x: int(x.split('.')[1]))

df['day_sin'] = np.sin(df.day*(2.*np.pi/30))
df['day_cos'] = np.cos(df.day*(2.*np.pi/30))

df['month_sin'] = np.sin(df.day*(2.*np.pi/12))
df['month_cos'] = np.cos(df.day*(2.*np.pi/12))
```

Оставим только записи с периодичностью раз в день, переведем во `float` все значения колонки `VV`, удалим оставшиеся промежуточные или уже неактуальные признаки и уберем признак `P`, который 100% коррелирует с признаком `P0` (оба признака об уровне влажности, но один линейно выражается через другой). Оставшиеся категориальные признаки закодируем через OneHotEncode с дропом первой колонки (т.к. бинарные значения k-1 колонок однозначно выражают k колонок)

```python
# delete everything but 1 record per day
df['Hour'] = df.Time.apply(lambda s: s.split()[-1])
df = df.drop(df[df.Hour != "12:20"].index)

month = df['month']
df = df.drop(columns=['day', 'month', 'Hour', 'Time'])
```

```python
# floatify VV
df.VV = df.VV.apply(lambda s: float(s.split()[0]))
```

```python
# since p and p0 are 100% correlated we drop them
df = df.drop(columns=['P'])
```

```python
# one hot encoding
df_encoded = pd.get_dummies(df, columns=['DD', 'c', 'WW'], drop_first=True)
```

### Уменьшение размерности признаков

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
# firstly you need to normalize features
scaler = StandardScaler()
scaled_matrix = scaler.fit_transform(df_encoded)
```

Для начала построим PCA для всех признаков, чтобы на графике каменистой осыпи посмотреть, какую часть всей информации (объясненнная дисперсия) отражает разное количество признаков

```python
default_pca = PCA()
default_pca.fit(scaled_matrix)
```

```python
PC_values = np.arange(default_pca.n_components_) + 1
plt.figure(figsize=(15,7))

plt.bar(PC_values, default_pca.explained_variance_ratio_)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.xticks(range(0,35))
plt.show()
```

```python
PC_values = np.arange(default_pca.n_components_) + 1
plt.figure(figsize=(15,7))
plt.plot(PC_values, default_pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.xticks(range(0,35))
plt.show()
```

```python
n_components = 24
explained_var = sum(default_pca.explained_variance_ratio_[:n_components])

print(f'Примерно {n_components} признака объясняет {explained_var} дисперсии всех данных')
```

```python
# PCA dimensions' reduction
pca_fit = PCA(n_components = n_components).fit_transform(scaled_matrix)
pca_fit.shape
```

```python
# t-sne dimensions' reduction
tsne_fit = TSNE(n_components=n_components, learning_rate='auto', init='random', 
                perplexity=3, method='exact', random_state=1).fit_transform(scaled_matrix)
tsne_fit.shape
```

### Сравнение двух алгоритмов

```python
pca_df = pd.DataFrame(pca_fit)
tsne_df = pd.DataFrame(tsne_fit)
```

```python
comparison = pd.DataFrame({'pca var': pca_df.std()**2, 't-sne var': tsne_df.std()**2, 
                           'pca mean': pca_df.mean(), 't-sne mean': tsne_df.mean()})
comparison
```

```python
pd.DataFrame(comparison.mean(), columns=['mean value for all features'])
```

Попробуем еще раз уменьшить все данные уже до 3 измерений каждым алгоритм и посмотреть, как они располагаются в пространстве

```python
pca3d = PCA(n_components = 3).fit_transform(pca_fit)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(pca3d[:, 0], pca3d[:, 1], pca3d[:, 2])
plt.title('PCA 3d projection')

plt.show()
```

```python
tsne3d = TSNE(n_components=3, learning_rate='auto', init='random', perplexity=3, random_state=1, method='exact').fit_transform(scaled_matrix)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(tsne3d[:, 0], tsne3d[:, 1], tsne3d[:, 2])
plt.title('t-SNE 3d projection')

plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(pca3d[:, 0], pca3d[:, 1], pca3d[:, 2], c=month)
plt.title('PCA 3d projection (month clusters)')

plt.show()
```
